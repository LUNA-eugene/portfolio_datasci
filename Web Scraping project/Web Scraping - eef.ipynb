## Web Scraping - Beautifulsoup
web scraping news from **eef**

```python
Title = []
Release_date = []
news_content = []
all_links = []

n_page = 1
while n_page <= 2:
    links = []
    eef_url = f'https://www.eef.or.th/tag/%E0%B9%80%E0%B8%94%E0%B9%87%E0%B8%81%E0%B8%AB%E0%B8%A5%E0%B8%B8%E0%B8%94%E0%B8%88%E0%B8%B2%E0%B8%81%E0%B8%A3%E0%B8%B0%E0%B8%9A%E0%B8%9A%E0%B8%81%E0%B8%B2%E0%B8%A3%E0%B8%A8%E0%B8%B6%E0%B8%81%E0%B8%A9/page/{n_page}/'
    print(eef_url)
    html = requests.get(eef_url)
    eef_html = soup(html.content ,"lxml")
    for link_news in eef_html.findAll('h2',{'class':'entry-title'}):
        links.append(link_news.a['href'])
    
    for link in links:
        print(link)
        all_links.append(link)
        
        page = requests.get(link)
        data = soup(page.content)
        for i in data('body') :
            Title.append(i.find('div',attrs= {'class':'title'}).text)
            Release_date.append(i.find('span',attrs={'class':'posted-on _heading'}).find('a').find('time',{"class":"entry-date published"}).text.strip())
            news_content.append(i.find('div',attrs={'class':'content-area'}).get_text(strip=True))
            
            
        
    print('Complete page number: ', n_page ,'จำนวนข่าวทั้งหมด: ', len(Title),' ข่าว')
    n_page += 1
 
```

```python
data = {'Title': Title,
        'Release_date':Release_date,
        'news_content':news_content,
        'links': all_links}

data = pd.DataFrame(data)
data
```

```python
# check columns name
data.columns
#check data type
data.dtypes
#check missing values
data.isna().sum().sort_values(ascending = False)
```

```python
# ลบพวกตัวแปลกๆ \xa0 , \u200b
data['news_content']= data['news_content'].replace(to_replace='\xa0', regex=True, value='')
```

```python
data['news_content']= data['news_content'].replace(to_replace='\u200b', regex=True, value='')
```

```python
# check
data['news_content'][7]
```

```python
# Split Date แตก Date ให้เป็น Day, Months, Years จาก spacebar
data['date_split'] = data['Release_date'].str.split('\s')
```

```python
# Create Columns Days , Months , Years
data['D'] = data['date_split'].str[0]
data['M'] = data['date_split'].str[1]
data['Y']  = data['date_split'].str[2]
```

```python
# ดูว่ามีเดือนอะไรบ้าง
data['M'].unique()
```

```python
# เปลี่ยนเดือนที่เป็นภาษาไทยให้เป็นตัวเลข เพื่ออยู่ในรูปแบบมาตรฐาน ISO
data['M'].replace({
    'มกราคม':'01',
    'กุมภาพันธ์':'02',
    'มีนาคม' :'03',
    'เมษายน':'04',
    'พฤษภาคม':'05',
     'มิถุนายน':'06',
     'กรกฎาคม':'07',
    'สิงหาคม':'08',
    'กันยายน':'09',
    'ตุลาคม':'10',
    'พฤศจิกายน':'11',
    'ธันวาคม': '12',
    }, inplace=True)
```

```python
data['Y'].unique
```

```python
data['Y'].replace({
    '2565':'2022',
    '2564':'2021',
    }, inplace=True)
```

```python
data
```

```python
# Create DATE Columns for Date ISO
# YYYY-MM-DD
data['Release_Date'] = data['Y']+'-'+data['M']+'-'+data['D']
```

```python
data.drop(['Release_date', 'date_split', 'D','M','Y'], axis=1, inplace=True)
```

```python
# Change format, datatype to datetime
data['Release_Date'] = pd.to_datetime(data['Release_Date'])
```

```python
data.to_csv('news_2.csv' , index=False)
```
