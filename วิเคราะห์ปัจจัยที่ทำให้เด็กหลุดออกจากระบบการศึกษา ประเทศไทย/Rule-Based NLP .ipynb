# Rule-Based NLP 

จากชุดข้อมูลเด็กหลุดออกจากระบบการศึกษาที่ได้มาจากฐานข้อมูลเพื่อความเสมอภาคทางการศึกษาและโครงการพาน้องกลับมาเรียน จำนวนทั้งสิ้น 5,298 records สามารถนำข้อมูลมาวิเคราะห์เพื่อจำแนกเด็กยากจนและคาดว่าไม่ยากจนได้จากการพิจารณา Keyword ในคอลัมน์สาเหตุการออก ความต้องการ และความคิดเห็นเพิ่มเติม ผ่านการใช้เทคนิค Rule-Based NLP ในการเขียนเงื่อนไขเพื่อจำแนกเด็กที่คาดว่ายากจน 

```python
# import pandas lib as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
 # read by default 1st sheet of an excel file
df = pd.read_excel('main_data.xlsx')
df
```

## Prepare Data

```python
# Columns with missing values
pd.DataFrame(df.isna().sum() , columns=['missing values'])
```

```python
# ตรวจสาเหตุหลักที่ทำให้เด็กออกจากระบบ
pd.DataFrame(df['สาเหตุหลัก'].value_counts())
```

```python
# Drop Record Students dead ลบนักเรียนที่เสียชีวิต
student_dead = df[df['สาเหตุหลัก'] == 'นักเรียนเสียชีวิต']
df.drop(student_dead.index , inplace=True , axis = 0)
```

```python
# Missing Value add - เพิ่มข้อมูลที่ขาดหายไปด้วย -
df.fillna("-", inplace = True)
```

```python
# จับกลุ่มสาเหตุการออกตามสาเหตุหลัก และนับจำนวน
df.groupby('สาเหตุหลัก')['สาเหตุการออก'].value_counts().to_frame()
```

## Rule Based
เราต้องการพิสูจน์ว่าความจนเป็นปัจจัยที่ทำให้เด็กหลุดออกจากระบบการศึกษามากที่สุด เลยตั้งเงื่อนไขว่าหาก 
- **สาเหตุการออก** คือ "ผู้ปกครองมีรายได้น้อย ไม่พอเพียง" แสดงว่าเด็กคนนี้มีภาวะยากจน 

แต่นอกจากสาเหตุการออกนี้ ในสาเหตุอื่นๆ ก็สามารถมีภาวะความยากจนแฝงอยู่ได้ เราเลยตั้งเงื่อนไขอื่นเพิ่มเติมได้แก่ 

- **ความต้องการ** หากนักเรียนระบุความต้องการที่สื่อถึงความยากจนเช่น ต้องการทุนการศึกษา, ลำบาก , ยากจน เป็นต้น แสดงว่าเด็กคนนี้มีภาวะยากจนหรือหาก
- **ความเห็นเพิ่มเติม** มี Keyword ที่สื่อถึงความยากจนเช่น ขาดแคลน , ไม่มั่นคง , ทำไรทำนา , เลี้ยงวัว เป็นต้น แสดงว่าเด็กคนนี้มีภาวะยากจน และหากไม่เข้าเงื่อนไขทั้งหมดแสดงว่าข้อมูลอาจจะไม่เพียงพอจึงคาดว่าไม่ยากจน

```python
# Create Condition to find จน เขียนเงื่อนไขที่เข้ากับคำว่ายากจน
conditions = [(df['สาเหตุการออก'] == 'ผู้ปกครองมีรายได้น้อย ไม่พอเพียง') | 
            (df['ความต้องการ'].str.contains('ทุน|เงิน|จน|รายได้|อาชีพ|ทำงาน|ประกอบอาชีพคนเดียว|ค่าใช้จ่าย|หางาน|งบประมาณ|ลำบาก|ช่วยเหลือในด้านปัจจัยพื้นฐาน|ต้องการให้หน่วยงานที่เกี่ยวข้อง มาดูแล พูดคุย ให้ความช่วยเหลือ|หน่วยงานที่เกี่ยวข้องเข้ามาให้ความช่วยเหลือ|ช่วยเหลือทางด้านปัจจัยในการดำรงชีวิตประจำวัน|อยากให้ช่วยเหลือชำระค่าเทอมในส่วนคงค้าง|หาเลี้ยง|อุปกรณ์การเรียน|ดำรงชีพ|ไม่มีผู้ปกครองช่วยดูแล|ค่าเล่าเรียน|งานทำ|ให้การดูแลอย่างใกล้ชิด|ค่าครองชีพ|ได้เรียนต่อ|ต้องการเรียนต่อ|สนับสนุนการศึกษา' ,case=True , na=False))|
            (df['ความเห็นเพิ่มเติม'].str.contains('ทุน|เงิน|จน|รายได้|อาชีพ|ทำงาน|ประกอบอาชีพคนเดียว|ค่าใช้จ่าย|ไม่มั่นคง|ขาดแคลน|ช่วยเหลือ|ลำบาก|รับจ้าง|ช่วยครอบครัว|หาเลี้ยง|ช่วยงาน|หางาน|ทำไร่|ทำนา|หางานทำได้แล้ว|ทำธุรกิจครอบครัว|ทำสวน|ครอบครัวไม่พร้อม|ไม่มีัญญาณโทรศัพท์|ไปไร่|ลูกจ้าง|ไม่มีใครคอยดูแล|เกษตร|งบประมาณ|หางานทำ|ค้าขาย|เลี้ยงวัว|ช่วยพ่อแม่|ทำงานแล้ว|หารายได้|หาเลียงตนเอง' ,case=True , na=False))]
```

```python
# สร้างคอลัมน์ใหม่ที่แสดงว่านักเรียนคนนี้ยากจนหรือไม่
df['ภาวะยากจน'] = np.select(conditions , ['ยากจน'] , default = 'คาดว่าไม่ยากจน')
```

```python
df.head(5)
```

```python
# นับจำนวนเด็กยากจน และคาดว่าไม่ยากจน
df['ภาวะยากจน'].value_counts().to_frame()
```

```python
# แสดงจำนวนความคิดเห็นเพิ่มเติม ตามภาวะความยากจน
df.groupby('ภาวะยากจน')['ความเห็นเพิ่มเติม'].value_counts().to_frame()
```
